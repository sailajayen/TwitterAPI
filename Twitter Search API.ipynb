```python
user_screenname = []
user_location = []
user_name = []
user_tz = []
tweet_text = []
tweet_tm = []
```

```python
#print status
#print
for status in twitter.search(q="elon musk", result_type='popular',count=20)["statuses"]:
    user = status["user"]["screen_name"].encode('utf-8')
    text = status["text"].encode('utf-8')
    tm = status["created_at"]
    loc = status["user"]["location"]
    nm = status["user"]["name"].encode('utf-8')
    tz = status["user"]["time_zone"]
#tweet_df = tweet_df
    print
    print (user, "--", loc, "--", nm,"--", tm, ":", text)
    print
    user_screenname.append(user)
    user_location.append(loc)
    user_name.append(nm)
    user_tz.append(tz)
    tweet_text.append(text)
    tweet_tm.append(tm)
```

```python
import pandas as pd
import numpy as np

tweets = pd.DataFrame(list(zip(user_screenname, user_location, user_name, user_tz, tweet_text, tweet_tm))
                      , columns=['user_screenname', 'user_location', 'user_name', 'user_tz', 'tweet_text', 'tweet_tm'])
```
